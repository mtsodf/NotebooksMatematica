```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
n, i, p, k = symbols("n i p k")
init_printing()
```

```python
def plot_seq(ax, an, n, ninic, nend, limit=None):
    an_values = []

    nvalues = range(ninic, nend+1)

    for j in nvalues:
        an_value = an.subs(n, j).evalf()
        an_values.append(an_value)
    
    ax.plot(nvalues, an_values, label="$%s$" % latex(an))
    
    if limit is not None:
        ax.plot([ninic, nend], [limit, limit], ls="--", color="black")

    ax.set_xlabel("$%s$" % latex(n), fontsize=15)
```

## Questão 1


### item A

```python
an = Sum(1/((i+1)*(i+2)), (i, 0, n))
an
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16,9))
plot_seq(ax, an, n, 0, 50, limit=1.0)   

ax.legend(fontsize=16)

plt.show()
```

### item B

```python
an = Sum(1/(i*(i+1)*(i+2)), (i, 1, n))
an
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16,9))
plot_seq(ax, an, n, 0, 50, limit=0.25)   

ax.legend(fontsize=16)

plt.show()
```

### item C

```python
an = Sum(1/((i+k)*(i+1+k)), (i, 0, n))
an
```

```python
an.subs(k, 1)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16,9))

for kvalue in [0.25, 0.5, 1.0, 2.0]:
    plot_seq(ax, an.subs(k, kvalue), n, 0, 30, limit=1/kvalue)   

    
ax.legend(fontsize=16)

plt.show()
```
