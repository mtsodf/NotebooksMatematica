```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
n, i, p, k = symbols("n i p k")
init_printing()
```

```python
def plot_seq(ax, an, n, ninic, nend, limit=None, scatter=False):
    an_values = []

    nvalues = range(ninic, nend+1)

    for j in nvalues:
        an_value = an.subs(n, j).evalf()
        an_values.append(an_value)
    
    if scatter:
        ax.scatter(nvalues, an_values, label="$%s$" % latex(an))
    else:
        ax.plot(nvalues, an_values, label="$%s$" % latex(an))
    
    if limit is not None:
        ax.plot([ninic, nend], [limit, limit], ls="--", color="black")

    ax.set_xlabel("$%s$" % latex(n), fontsize=15)
```

```python
an = Sum(1/(i*(i+1)*(i+2)*(i+3)), (i, 1, n))
an
```

```python
bn = Rational(1,3) * (Rational(1,6) - 1/((n+1)*(n+2)*(n+3)))
bn
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16,9))
plot_seq(ax, an, n, 0, 20, limit=1.0/18)
plot_seq(ax, bn, n, 0, 20, limit=1.0/18, scatter=True) 

ax.legend(fontsize=16)

plt.show()
```
