```python
n, i, p = symbols("n i p")
init_printing()
```

```python
an = 1/(n**2)
an = n**(1/(n*n))
an = (1-1/n)**n
an = 1/(n*n-1)
an = parse_expr("1 / log(n+1)**n") 
an = parse_expr("(-1)**(n+1)/(sqrt(n+1) - sqrt(n))")
an = Product(2*i-1, (i, 1, n)) / Product(2*i, (i, 1, n))
an = (log(n))/sqrt(n)
an = exp(1/n) / n**2
an = sqrt(n)/(n+5)
an = 1/(log(n)**log(n))
an = 2**(1/n) - 1
an = 1/(n*(log(n)**i))
an = 1/n**2
an = 1/ (n * n * log(n)**p)
qtd = 1000
inic = 2
```

```python
sn = Sum(an.subs(n, i), (i, inic, n))
sn.subs(p, 1)
```

```python
L = limit_seq(sn.subs(p, 1))
```

```python
an_values = []
sn_values = []
sn_values_mine = []
for j in range(inic, qtd):
    if i%100 == 0:
            print("Calculando %d" % i)
    an_value = an.subs(n, j).evalf()
    
    an_values.append(an_value)
    
    sn_values.append(sn.subs(n, j).evalf())
    
    if j == inic:
        sn_values_mine.append(an_value)
    else:
        sn_values_mine.append(an_value + sn_values_mine[-1])
```

```python
fig, ax1 = plt.subplots(figsize=(16,9))

ax1.plot(np.arange(inic, len(an_values) + inic), an_values, color="blue", label="Sequencia")

ax2 = ax1.twinx()

#ax2.scatter(np.arange(inic, len(an_values) + inic), sn_values, color="red", label="Serie")
ax2.semilogy(np.arange(inic, len(an_values) + inic), sn_values, color="red", label="Serie", marker="o", lw=0)
#ax2.plot(np.arange(inic, len(an_values) + inic), sn_values_mine, color="green", label="Serie")
ax2.semilogy(np.arange(inic, len(an_values) + inic), sn_values_mine, color="green", label="Serie")



ax1.legend()
ax2.legend()
plt.show()
```

```python
sn_values = []
for i in range(inic, 1000):
    if i%100 == 0:
            print("Calculando %d" % i)
    
    
    an_value = an_values[i-inic]
    an_value = an_value*an_value*an_value
    if i == inic:
        sn_values.append(an_value)
    else:
        sn_values.append(an_value + sn_values[-1])
```

```python
fig, ax1 = plt.subplots(figsize=(16,9))

ax1.plot(np.arange(inic, len(an_values) + inic), an_values, color="blue", label="Sequencia")

ax2 = ax1.twinx()

ax2.scatter(np.arange(inic, len(an_values) + inic), sn_values, color="red", label="Serie")

ax1.legend()
ax2.legend()
plt.show()
```
